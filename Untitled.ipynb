```python
#Visulaization through seaborn  on iris dataset
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr());
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,palette='rainbow',kind='hex')
```

```python
sns.jointplot(x='sepal_width',y='petal_width',kind='reg',data=iris)
```

### Pair plot

```python
sns.pairplot(iris);
```

```python
sns.pairplot(iris,hue='species')
```

### displot


```python
sns.displot(iris['sepal_length'])
```

```python
sns.displot(iris['sepal_length'],kde=False,bins=10)
```

### Categorical Plots

```python
#countplot
sns.countplot(x='species',data=iris);
```

```python
iris['species'].value_counts()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
```

```python
sns.boxplot(x='species',y='sepal_width',data=iris)
```

```python
sns.violinplot(x='sepal_length',y='species',data=iris)
```
